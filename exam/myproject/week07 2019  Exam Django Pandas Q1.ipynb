```python
%matplotlib inline
import pandas as pd
```

```python
df_rent = pd.DataFrame(list(Rent.objects.all().values()))
df_rent.index = df_rent['id']
df_rent.drop(['id'], axis=1, inplace=True)
df_rent.head()
```

```python
df_room = pd.DataFrame(list(Room.objects.all().values()))
df_room.index = df_room['id']
df_room.drop(['id'], axis=1, inplace=True)
df_room.head()
```

```python
df_member = pd.DataFrame(list(Member.objects.all().values()))
df_member.index = df_member['id']
df_member.drop(['id'], axis=1, inplace=True)
df_member.head()
```

```python
df=df_rent\
    .merge(df_room, left_on='room_id',right_on='id')\
    .merge(df_member, left_on='member_id',right_on='id')
df.head()
```

```python
df\
    .groupby(by=['member_id','name'])\
    .agg({'cost' : ['sum', 'count']})\
    .sort_values(('cost','sum'), ascending=False)
```
